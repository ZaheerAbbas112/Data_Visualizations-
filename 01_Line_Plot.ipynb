## Adding titles

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")

plt.show()
```

## Adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool
# draw a line plot
sns.lineplot(x="sepal_length", y="sepal_width", data=phool)
plt.title("Phoolon ka Plot")
plt.xlim(3)
plt.ylim(1.5)

plt.show()
```
